```python
from surprise import Dataset, Reader
from surprise.model_selection import train_test_split
from surprise import KNNBasic
from surprise import accuracy

data = Dataset.load_builtin('ml-100k')

reader = Reader(rating_scale=(1, 5))

dataset = data.build_full_trainset()

sim_options = {
    'name': 'cosine',
    'user_based': True
}
algo = KNNBasic(sim_options=sim_options)

algo.fit(dataset)

user_id = str(1)
user_ratings = dataset.ur[int(user_id)]
movies_rated_by_user = [item for item, _ in user_ratings]

all_movie_ids = list(dataset.ir.keys())

movies_to_predict = [movie_id for movie_id in all_movie_ids if movie_id not in movies_rated_by_user]

predictions = [algo.predict(user_id, movie_id) for movie_id in movies_to_predict]

predictions.sort(key=lambda x: x.est, reverse=True)

print("Top 5 recommended movies:")
for pred in predictions[:5]:

    movie_id = pred.iid

    movie_title = dataset.to_raw_iid(movie_id)
    print(f"{movie_title} (predicted rating: {pred.est:.2f})")
```
